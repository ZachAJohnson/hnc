### Define parameters and PLASMA

```python
ni_cc = 1e22
Te_eV = 1#5.56
Te_AU = Te_eV*eV_to_AU 
Ti_eV = Te_eV

Z, A = 1,1
Zbar = Z


name2 = 'OZ - th'
plasma_ozt = lambda T_eV, ni_cc: PLASMA(Z, A, ni_cc, T_eV, T_eV, Zbar=Z,
                 Î²u_options = {'add_bridge':False,'bridge':'ocp', 'pseudopotential':False}, 
                 qsp_options= {'r_c':0,'which_Tij':'thermal'},
                 hnc_options= {'kappa_multiscale':10, 'R_max':20,'N_bins':1000, 
                               'h_max':2e2,'oz_method':'standard'},
                 hnc_solve_options={'alpha_Picard':0.5, 'num_iterations':1e4})

plasma = plasma_ozt(Te_eV, ni_cc)
```

```python
fig, axs = plt.subplots(1,2, figsize=(16,6))

r_array = plasma.hnc.r_array

# Potentials
ax = axs[0]
ax.set_title(r"$\beta_{ei} u_{ei}$", fontsize=20)

ax.plot(r_array, plasma.qsp.Î²v_Deutsch(plasma.qsp.Îei, r_array, plasma.qsp.Îei), label="Deutsch")
ax.plot(r_array, plasma.qsp.Î²v_Kelbg(plasma.qsp.Îei, r_array, plasma.qsp.Îei), label="Kelbg")
ax.plot(r_array, plasma.qsp.Î²vei(r_array), label="Improved (Filinov fit) Kelbg")

#Forces
ax = axs[1]
ddr = lambda f: np.gradient(f, r_array*plasma.qsp.ri)
ax.set_title(r"$\nabla\beta_{ei} u_{ei}$", fontsize=20)

ax.plot(r_array, ddr(plasma.qsp.Î²v_Deutsch(plasma.qsp.Îei, r_array, plasma.qsp.Îei)), label="Deutsch")
ax.plot(r_array, ddr(plasma.qsp.Î²v_Kelbg(plasma.qsp.Îei, r_array, plasma.qsp.Îei)), label="Kelbg")
ax.plot(r_array, ddr(plasma.qsp.Î²vei(r_array)), label="Improved (Filinov fit) Kelbg")

#Kato:
ax.plot(r_array[:30], 2*Z*np.ones_like(r_array)[:30],'k--', label='Kato Cusp Condition')
# Configure axes
for ax in axs:
    ax.set_yscale('symlog', linthresh=0.01)
    ax.set_xscale('log')

    ax.legend(fontsize=14)

```
