# Integral Equation Model Comparisons
Purpose of this notebook is to EOS calculations for comparisons to Triola paper

[DOI: https://doi.org/10.1063/5.0110725](https://doi.org/10.1063/5.0110725)

```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv

from hnc.hnc.PIE import Plasma_of_Ions_and_Electrons as PLASMA

from hnc.hnc.constants import *
from hnc.hnc.misc import n_from_rs, rs_from_n

prop_cycle = plt.rcParams['axes.prop_cycle']
colors = prop_cycle.by_key()['color']
                
from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}

</style>
""")
```

### Set Parameters in Paper and Plot

```python
name = 'Hydrogen'
Z, A = 1,1 

Te_eV = 1000 # eV
Ti_eV = Te_eV
ni_cc_list = np.array([5.98e23])#np.geomspace(1e20,1e23,num=10)#n_from_rs(ri_cm)

ri_cm_list = rs_from_n(ni_cc_list)  # cm
ri_AU_list = ri_cm_list*cm_to_AU    # AU 

```

```python
def plot_deviation(n_cc_list, U_dev_list, ax=None, label='', name=''):
    if ax==None:
        fig, ax = plt.subplots(figsize=(8,6))
    
    ax.plot(n_cc_list, U_dev_list, label=label)
    
    ax.set_xlabel(r"$n_i$ [cc$^{-1}$]", fontsize=20)
    ax.set_ylabel(name, fontsize=20)
    ax.set_xscale('log')
    ax.set_ylim(-0.5,0)
    
```

#### Run Ornstein-Zernike Closure

```python
# c_s_k_guess = plasma_oz.hnc.c_s_k_matrix.copy()
```

```python
def plasma_output(plasma):
    plasma.run_hnc()
    Uid = plasma.hnc.ideal_energy_density()
    Uex = plasma.hnc.excess_energy_density()
    Pid = plasma.hnc.ideal_pressure()
    Pex = plasma.hnc.excess_pressure()
    return Uid, Uex, Pid, Pex

U_list_1 = []
P_list_1 = []
name1 = 'OZ - g'
plasma1 = lambda ni_cc: PLASMA(Z, A, ni_cc, Ti_eV, Te_eV, Zbar=Z,
                 βu_options = {'add_bridge':False,'bridge':'ocp', 'pseudopotential':False}, 
                 qsp_options= {'r_c':0,'which_Tij':'geometric'},
                 hnc_options= {'kappa_multiscale':10, 'R_max':20,'N_bins':1000, 
                               'h_max':2e2,'oz_method':'standard'},
                 hnc_solve_options={'alpha_Picard':0.5, 'num_iterations':1e4})
U_list_2 = []
P_list_2 = []
name2 = 'OZ - th'
plasma2 = lambda ni_cc: PLASMA(Z, A, ni_cc, Ti_eV, Te_eV, Zbar=Z,
                 βu_options = {'add_bridge':False,'bridge':'ocp', 'pseudopotential':False}, 
                 qsp_options= {'r_c':0,'which_Tij':'thermal'},
                 hnc_options= {'kappa_multiscale':10, 'R_max':20,'N_bins':1000, 
                               'h_max':2e2,'oz_method':'standard'},
                 hnc_solve_options={'alpha_Picard':0.5, 'num_iterations':1e4})
U_list_3 = []
P_list_3 = []
name3 = 'OZ - g  Pauli'
plasma3 = lambda ni_cc: PLASMA(Z, A, ni_cc, Ti_eV, Te_eV, Zbar=Z, find_βuee=True,
                 βu_options = {'add_bridge':False,'bridge':'ocp', 'pseudopotential':False}, 
                 qsp_options= {'r_c':0,'which_Tij':'geometric'},
                 hnc_options= {'kappa_multiscale':10, 'R_max':20,'N_bins':1000, 
                               'h_max':2e2,'oz_method':'standard'},
                 hnc_solve_options={'alpha_Picard':0.5, 'num_iterations':1e4})
U_list_4 = []
P_list_4 = []
name4 = 'SVT - th'
plasma4 =  lambda ni_cc: PLASMA(Z, A, ni_cc, Ti_eV, Te_eV, Zbar=Z,
                 βu_options = {'add_bridge':False,'bridge':'ocp', 'pseudopotential':False}, 
                 qsp_options= {'r_c':0},
                 hnc_options= {'kappa_multiscale':10, 'R_max':20,'N_bins':1000, 
                               'h_max':2e2,'oz_method':'svt'},
                 hnc_solve_options={'alpha_Picard':0.5,'num_iterations':1e4})

P_hnc_to_bar = lambda P_hnc: (P_hnc/ri_AU_list[0]**3) *  AU_to_bar 
P_hnc_to_Mbar = lambda P_hnc: 1e-6*P_hnc_to_bar(P_hnc)
P_hnc_to_GPa = lambda P_hnc:  P_hnc_to_bar(P_hnc) * 1e5 / 1e9 
```

### Compute all plasma models for list of densities

```python

for ni_cc in ni_cc_list:
    #1
    plasma = plasma1(ni_cc)
    Uid, Uex, Pid, Pex = plasma_output(plasma)
    U_list_1.append( Uex/Uid )
    P_list_1.append( Pex/Pid )
    #2
    plasma = plasma2(ni_cc)
    Uid, Uex, Pid, Pex = plasma_output(plasma)
    U_list_2.append( Uex/Uid )
    P_list_2.append( Pex/Pid )
    #3
    plasma = plasma3(ni_cc)
    Uid, Uex, Pid, Pex = plasma_output(plasma)
    U_list_3.append( Uex/Uid )
    P_list_3.append( Pex/Pid )
    #4
    plasma = plasma4(ni_cc)
    Uid, Uex, Pid, Pex = plasma_output(plasma)
    U_list_4.append( Uex/Uid )
    P_list_4.append( Pex/Pid )
    
```

Now Plot 

```python
fig, axs = plt.subplots(1,2, figsize=(16,6))
fig.suptitle(r"$T_i=T_e = 20 $ eV", fontsize=25, y=0.9)

# Plot Energy ratio
plot_deviation(ni_cc_list, U_list_1, ax=axs[0], label=name1,name=r"$U_{ex}/U_{id}$")
plot_deviation(ni_cc_list, U_list_2, ax=axs[0], label=name2,name=r"$U_{ex}/U_{id}$")
# plot_U_deviation(ni_cc_list, U_list_3, ax=ax, label=name3)
plot_deviation(ni_cc_list, U_list_4, ax=axs[0], label=name4,name=r"$U_{ex}/U_{id}$")

# Plot Pressure ratio
plot_deviation(ni_cc_list, P_list_1, ax=axs[1], label=name1, name=r"$P_{ex}/P_{id}$")
plot_deviation(ni_cc_list, P_list_2, ax=axs[1], label=name2, name=r"$P_{ex}/P_{id}$")
plot_deviation(ni_cc_list, P_list_4, ax=axs[1], label=name4, name=r"$P_{ex}/P_{id}$")

#plot 'Data' from linear theory in Triola paper
Triola_Udata1 = read_csv("Triola_Ti=Te=20eV_H-U.txt", delim_whitespace=True, header=2)
Triola_Pdata1 = read_csv("Triola_Ti=Te=20eV_H-P.txt", delim_whitespace=True, header=2)

axs[0].plot(Triola_Udata1['ni[1/cc]'], Triola_Udata1['u/uid-1'],'k--' ,label='Linear (Triola 2022)')
axs[1].plot(Triola_Pdata1['ni[1/cc]'], Triola_Pdata1['p/pid-1'],'k--' ,label='Linear (Triola 2022)')
for ax in axs:
    ax.legend(fontsize=20, loc='lower left')
plt.tight_layout()
```
