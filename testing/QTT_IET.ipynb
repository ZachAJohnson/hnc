```python
Γ, κ = 10, 3
n0 = 3/(4*π)

βu_r_func = lambda r: Γ/r*np.exp(-κ*r)

N_bits = 12
N_grid = 2**N_bits
print(f"N_points = {N_grid}")
eps = 1e-4
R_max = 100

def qtt_from_tensor(tensor, eps=1e-6, order='F'):
    return ttpy.vector(tensor_to_binarytensor(tensor, order=order), eps=eps )

r_array = np.linspace(1e-5, R_max, num=N_grid)
k_array = π*(np.arange(N_grid)+0.5)/R_max

r_qtt = qtt_from_tensor(r_array, order='F', eps=eps)
k_qtt = qtt_from_tensor(k_array, order='F', eps=eps)

invr_qtt = qtt_from_tensor(1/r_array, order='F', eps=eps)
invk_qtt = qtt_from_tensor(1/k_array, order='F', eps=eps)

βu_r_array = Γ/r_array*np.exp(-κ*r_array)
βu_r_bt = tensor_to_binarytensor(βu_r_array, order='F')
βu_r_qtt = ttpy.vector(βu_r_bt, eps=eps)

f_M_r_array = np.exp(-βu_r_array) - 1
f_M_r_bt = tensor_to_binarytensor(f_M_r_array, order='F')
f_M_r_qtt = ttpy.vector(f_M_r_bt, eps=eps)
```

```python
γ_s_qtt_list = []
resnorm_list = []

from scipy.optimize import minimize

def print_minimize_output(sol):
    print(f"\tfun: {sol.fun:0.3e} \n\tmessage: {sol.message}, \n\tnfev: {sol.nfev}, \n\tnit: {sol.nit}, \n\tsuccess: {sol.success}")#, \n\tx: {sol.x}")
    
def update_qtt_with_flatcore(qtt, i, flatcore):
    qtt_cores = get_tt_cores(qtt)
    qtt_core_i_shape = qtt_cores[i].shape
    qtt_cores[i] = flatcore.reshape(qtt_core_i_shape)
    return ttpy.vector.from_list(qtt_cores)
    
def qtt_corebycore_optimizer(γ_r_qtt, tol=1e-1):
    γ_cores = get_tt_cores(γ_r_qtt)
    
    for i in np.arange(N_bits)[::-1]:
        γ_core_i = γ_cores[i]
        γ_core_i_flat = γ_core_i.flatten()
        def get_residual_norm(flat_core):
            new_γ_r_qtt = update_qtt_with_flatcore(γ_r_qtt, i, flat_core)
            guess_γ_r_qtt = iet.get_γ_r_qtt(iet.get_h_r_qtt_from_γ_qtt(new_γ_r_qtt), iet.get_c_r_qtt_from_γ_qtt(new_γ_r_qtt) )
            res_norm = Fr_err(guess_γ_r_qtt, new_γ_r_qtt)
            return res_norm # inefficient full
        options={
        'xatol': tol,    # stop if x changes less than this
        'fatol': tol,    # function tolerance
        'maxiter': 10000}
        
        min_sol = minimize(get_residual_norm, γ_core_i_flat, method='Nelder-Mead',options=options)
        
        print(f"\nCore {i}/{N_bits-1}: {γ_core_i.shape}")
        print_minimize_output(min_sol)
        
        γ_r_qtt = update_qtt_with_flatcore( γ_r_qtt, i, min_sol.x)
        γ_s_qtt_list.append(γ_r_qtt)
        resnorm_list.append(min_sol.fun)
    return γ_r_qtt
    
γ_r_qtt_init = iet.get_γ_r_qtt(iet.f_M_r_qtt, iet.f_M_r_qtt)
γ_r_sol1 = qtt_corebycore_optimizer( γ_r_qtt_init, tol=1e-1)    
γ_r_sol2 = qtt_corebycore_optimizer( γ_r_sol1, tol=1e-2)    
# γ_r_sol3 = qtt_corebycore_optimizer( γ_r_sol2, tol=1e-3)    
```

```python
from scipy.optimize import minimize


def get_flat_qtt(qtt):
    cores = get_tt_cores(qtt)
    flat_qtt_list = []
    for i, core in enumerate(cores):
        flat_qtt_list = flat_qtt_list + list(core.flatten())
    return flat_qtt_list

def reform_qtt_from_flat(flat_qtt, original_shapes_list):
    start_index = 0
    cores = []
    for i, shape in enumerate(original_shapes_list):
        size = np.prod(shape)
        end_index = start_index+size
        cores.append(np.array(flat_qtt[start_index:end_index]).reshape(shape))
        start_index=end_index
    # Reshape using original_shapes_list:
    return ttpy.vector.from_list(cores) # the list of reshaped 3D arrays

γ_s_qtt_list = []
resnorm_list = []

def qtt_full_optimizer(γ_r_qtt):
    γ_flat_qtt_init = get_flat_qtt(γ_r_qtt)    
    original_qtt_shapes = get_tt_core_shapes(γ_r_qtt)
    def get_residual_norm(flat_qtt):
        new_γ_r_qtt = reform_qtt_from_flat(flat_qtt, original_qtt_shapes)
        guess_γ_r_qtt = iet.get_γ_r_qtt(iet.get_h_r_qtt_from_γ_qtt(new_γ_r_qtt), iet.get_c_r_qtt_from_γ_qtt(new_γ_r_qtt) )
        res_norm = Fr_err(guess_γ_r_qtt, new_γ_r_qtt)
        print(f"\t|norm|={res_norm}")
        γ_s_qtt_list.append(new_γ_r_qtt)
        resnorm_list.append(res_norm)
        return res_norm # inefficient full
    min_sol = minimize(get_residual_norm, γ_flat_qtt_init, tol=1e-1)
    print(f"\nCore {i}/{N_bits}: ", min_sol)
    γ_r_qtt = update_qtt_with_flatcore( γ_r_qtt, i, min_sol.x)
    return γ_r_qtt
    
γ_r_qtt_init = iet.get_γ_r_qtt(iet.f_M_r_qtt, iet.f_M_r_qtt)
qtt_full_optimizer( γ_r_qtt_init)    
```
