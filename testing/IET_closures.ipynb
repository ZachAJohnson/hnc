```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_csv

# from hnc.hnc.hnc import Integral_Equation_Solver as IET
from hnc.hnc.hnc import Integral_Equation_Solver as old_IET
from hnc.hnc.constants import *

from pandas import read_csv
from scipy.fft import fft, ifft, fftshift, ifftshift, dct, dst
                
plt.rcParams.update({'font.size': 10})
plt.rcParams["figure.figsize"] = (5,3.5)
plt.rcParams["figure.dpi"] = 200
plt.rcParams['xtick.labelsize']=16
plt.rcParams['ytick.labelsize']=16
plt.rcParams['ytick.major.size']=5
plt.rcParams['ytick.minor.size']=3
plt.rcParams['xtick.major.size']=5
plt.rcParams['xtick.minor.size']=3
plt.rcParams['legend.fontsize'] = 10
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['text.usetex'] = True

color_cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
# Γ, κ = 10, 3
n0 = 3/(4*π)

βu_r_func = lambda r: Γ/r*np.exp(-κ*r)

N_bits = 14
N_grid = 2**N_bits
print(f"N_points = {N_grid}")
eps = 1e-4
R_max = 100
```

```python
Γ_list = [1,10,100,500]
κ = 3
iet_list = []
iet_PY_list = []
for Γ in Γ_list:
    iet = old_IET(1, [[Γ]], [[n0]], [[1]], [1], kappa = κ, dst_type=4, N_bins = N_grid, R_max = R_max, closure='HNC')
    iet_PY = old_IET(1, [[Γ]], [[n0]], [[1]], [1], kappa = κ, kappa_multiscale=0, dst_type=4, N_bins = N_grid, R_max = R_max, closure='PY')
    iet.HNC_solve(alpha_Picard=0.1, tol=1e-10, num_iterations=100000, iters_to_wait=1e4)
    iet_PY.c_k_matrix[:,:,:] = np.exp(-iet_PY.βu_k_matrix[:,:,:]) - 1
    iet_PY.c_r_matrix = iet_PY.FT_k_2_r_matrix(iet_PY.c_k_matrix)
    iet_PY.c_s_k_matrix = iet_PY.c_k_matrix + iet_PY.βu_l_k_matrix
    iet_PY.c_s_r_matrix = iet_PY.FT_k_2_r_matrix(iet_PY.c_s_k_matrix)
    iet_PY.HNC_solve(alpha_Picard=0.1, tol=1e-10, num_iterations=100000, iters_to_wait=1e4)
    iet_list.append(iet)
    iet_PY_list.append(iet_PY)
```

```python
fig, ax = plt.subplots()

for i, (Γ, iet, iet_PY) in enumerate(zip(Γ_list, iet_list, iet_PY_list)):
    if i<4:
        ax.plot(iet_PY.r_array, iet_PY.γs_r_matrix[0,0] + iet_PY.βu_l_r_matrix[0,0],':', color=color_cycle[i], label=rf"$\Gamma={Γ:0.3f}$ (PY)")
        
for i, (Γ, iet, iet_PY) in enumerate(zip(Γ_list, iet_list, iet_PY_list)):
    if i<4:
        ax.plot(iet.r_array, iet.γs_r_matrix[0,0] + iet.βu_l_r_matrix[0,0], color=color_cycle[i],alpha=0.5, label=rf"$\Gamma={Γ:0.3f} (HNC)$")

ax.set_xscale('log')
ax.set_yscale('symlog', linthresh=0.1)
# ax.set_yscale('log')
# ax.set_ylim(1e-10,None)
ax.text(5, 1, rf"$\kappa={κ:0.3f}$")
ax.legend(loc='lower left', fontsize=6, ncol=2)
ax.grid(alpha=0.3)
```
