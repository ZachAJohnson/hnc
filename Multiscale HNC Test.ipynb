```python
import numpy as np
import matplotlib.pyplot as plt

from hnc_n_multiscale import  HNC_solver
```

# One Species

```python
N_species = 1
Gamma = np.array(  [[100]])
names = ["Electron"] 
kappa = 0e-3
rho = np.array([3/(4*np.pi)])
hnc = HNC_solver(N_species, Gamma=Gamma, kappa=kappa, tol=1e-4,
                 kappa_multiscale=2, rho = rho, num_iterations=int(1e4), 
                 R_max=20, N_bins=1000, names=names)

hnc.HNC_solve(alpha=1)

h_r_matrix = hnc.h_r_matrix
c_r_matrix = hnc.c_r_matrix

# To get the radial distribution function g(r), you can add 1 to the h_r_matrix:
g_r_matrix = h_r_matrix + 1

# To visualize the results, you can use matplotlib to plot the g(r) function:
# hnc.plot_species((0,0))
# hnc.plot_species_convergence((0,0))
```

```python
fig, ax = plt.subplots(figsize=(10,6))

MM_r, MM_g = np.loadtxt("/home/zach/plasma/atomic_forces/average_atom/jupyter/Murillo_HNC.txt").T
ax.plot(MM_r[::3], MM_g[::3], 'k.', label="MM HNC- One Species")
ax.plot(hnc.r_array, hnc.h_r_matrix[0,0]+1,'--.', label="One Species")
ax.set_ylabel(r'$g(r/r_s)$',fontsize=20)
ax.set_xlabel(r'$r/r_s$',fontsize=20)
ax.set_ylim(0,2)
# ax.set_xlim(1900,2000)
# ax.set_xlim(0,10)
ax.tick_params(labelsize=15)
ax.legend(fontsize=15)
```

# Two Species

```python
N_species = 2
Gamma = np.array(  [[5,  0.5],
                    [0.5,  5]])

names = ["Ion-1", "Electron", ] 
kappa = 3
rho = np.array([  3/(4*np.pi), 5*3/(4*np.pi)  ])
hnc2 = HNC_solver(N_species, Gamma=Gamma, kappa=kappa, tol=1e-3, kappa_multiscale=1, rho = rho, num_iterations=int(1e4), R_max=10, N_bins=1000, names=names)

hnc2.HNC_solve(alpha=1e-1)
# print(hnc.c_r_matrix)

h_r_matrix = hnc2.h_r_matrix
c_r_matrix = hnc2.c_r_matrix

# To get the radial distribution function g(r), you can add 1 to the h_r_matrix:
g_r_matrix = h_r_matrix + 1

# To visualize the results, you can use matplotlib to plot the g(r) function:
# hnc.plot_species((0,0))
# hnc.plot_species_convergence((0,0))
hnc2.plot_g_all_species()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

MM_r, MM_g = np.loadtxt("/home/zach/plasma/atomic_forces/average_atom/jupyter/Murillo_HNC.txt").T
ax.plot(MM_r[::3], MM_g[::3], 'k.', label="MM HNC- One Species")
ax.plot(hnc.r_array, hnc.h_r_matrix[0,0]+1,'--.', label="One Species")
# ax.plot(hnc2.r_array, hnc2.h_r_matrix[0,0]+1, label="Two Species")
ax.set_ylabel(r'$g(r/r_s)$',fontsize=20)
ax.set_xlabel(r'$r/r_s$',fontsize=20)
ax.set_ylim(0,2)
# ax.set_xlim(1900,2000)
# ax.set_xlim(0,10)
ax.tick_params(labelsize=15)
ax.legend(fontsize=15)
```

# Three Species

```python
N_species = 3
Gamma = np.array(  [[1,  0.1,  3],
                    [0.1,  5,   3],
                    [3,    3,   5]])

names = ["Ion-1","Ion-2", "Electron", ] 
kappa = 3
rho = np.array([  3/(4*np.pi), 3/(4*np.pi), 3/(4*np.pi)  ])
hnc3 = HNC_solver(N_species, Gamma=Gamma, kappa=kappa, tol=1e-1, 
                  kappa_multiscale=1, rho = rho, num_iterations=int(1e4), R_max=10, N_bins=1000, names=names)

hnc3.HNC_solve()
# print(hnc.c_r_matrix)

h_r_matrix = hnc3.h_r_matrix
c_r_matrix = hnc3.c_r_matrix

# To get the radial distribution function g(r), you can add 1 to the h_r_matrix:
g_r_matrix = h_r_matrix + 1

# To visualize the results, you can use matplotlib to plot the g(r) function:
# hnc.plot_species((0,0))
# hnc.plot_species_convergence((0,0))
hnc3.plot_g_all_species()
```

```python
hnc3.h_r_matrix[0,0]/hnc.h_r_matrix[0,0]
```

```python
fig, ax = plt.subplots(figsize=(10,6))

MM_r, MM_g = np.loadtxt("/home/zach/plasma/atomic_forces/average_atom/jupyter/Murillo_HNC.txt").T
ax.plot(MM_r[::3], MM_g[::3], 'k.', label="MM HNC- One Species")
ax.plot(hnc.r_array, hnc.h_r_matrix[0,0]+1, label="One Species")
ax.plot(hnc3.r_array, hnc3.h_r_matrix[0,0]+1, label="Three Species")
ax.set_ylabel(r'$g(r/r_s)$',fontsize=20)
ax.set_xlabel(r'$r/r_s$',fontsize=20)
ax.set_ylim(0,2)
ax.tick_params(labelsize=15)
ax.legend(fontsize=15)
```

# Effective Potentials

```python
hnc2.invert_HNC( 1 , alpha=1, tol=1e-3)
```

```python
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(hnc2.r_array, hnc2.βueff_r_matrix[0,0], label="Effective")
ax.plot(hnc2.r_array, hnc2.βu_r_matrix[0,0], label="Normal")
ax.set_ylabel(r'$\beta u(r/r_s)$',fontsize=20)
ax.set_xlabel(r'$r/r_s$',fontsize=20)
# ax.set_ylim(,2)
ax.set_yscale('symlog',linthresh=1)
ax.tick_params(labelsize=15)
ax.legend(fontsize=15)
```
